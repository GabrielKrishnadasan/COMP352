```python
babyNamesDataFrame["name_lengths"]=babyNamesDataFrame["Name"].str.len()
```

```python
sorted_babyNamesDataFrame=babyNamesDataFrame.sort_values("name_lengths")
```

```python
sorted_babyNamesDataFrame.head()
```

```python
babyNamesDataFrame=babyNamesDataFrame.drop("name_lengths",axis=1)
```

```python
babyNamesDataFrame.head()
```

```python
#Add a column â€œa_occurrencesâ€ to the babyNames data frame with count of â€œaâ€ 
#character in the data frame.

#str.count('a')

babyNamesDataFrame["a_occurrences"]=babyNamesDataFrame["Name"].str.count('a')
```

```python
babyNamesDataFrame.head()
```

```python
def ea_and_dr_occurrences(string):
    return string.count("ea")+string.count("dr")

babyNamesDataFrame["ea_and_dr_occurrences"]=babyNamesDataFrame["Name"].map(ea_and_dr_occurrences)
```

```python
babyNamesDataFrame.head()
```

```python
# Find all Names with both "ea" or "dr" substrings

babyNamesDataFrame.query("ea_and_dr_occurrences>0")
```

```python
# Find all Names with both "ea" and "dr" substrings

babyNamesDataFrame.query("Name.str.contains('ea') and Name.str.contains('dr')")
```

```python
def ammd(series):
    return max(series)-min(series)
```
